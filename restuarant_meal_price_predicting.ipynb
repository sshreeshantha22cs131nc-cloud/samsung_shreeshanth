```python
X = df['Food_Quality']

Y = df['Price']
```

```python
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
```

```python
summary = model.summary()
print(summary)
```

```python
sm.stats.linear_rainbow(model)
```

```python
model.params
```

```python
model.rsquared
```
